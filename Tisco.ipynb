<span style="font-size:20px;">Data Exploration:
Calculated basic summary for the numeric columns (mean, median, min, max, standard deviation) </span>

```python
column_to_analyze = "Duration(in hrs)"
mean_value = coursera[column_to_analyze].mean()
median_value = coursera[column_to_analyze].median()
std_deviation = coursera[column_to_analyze].std()

print(f"\nBasic Statistics for '{column_to_analyze}':")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Standard Deviation: {std_deviation:.2f}")

min_value = coursera[column_to_analyze].min()
max_value = coursera[column_to_analyze].max()

print(f"min value: {min_value:.2f}")
print(f"max value: {max_value:.2f}")
```

<span style="font-size:20px;">Frequency plotting of Course Completion </span>

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 6))
ax = sns.countplot(data = coursera, x = 'Completed')
for bars in ax.containers:
    ax.bar_label(bars)

plt.suptitle('Frequency Distribution of Completion')
plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.show()
```
